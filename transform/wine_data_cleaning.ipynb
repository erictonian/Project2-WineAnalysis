```python
# import the dependencies
import pandas as pd
import numpy as np
```

```python
# import wine data into pandas
wine_data = "../original_data/winemag-data-130k-v2.csv"
wine_df = pd.read_csv(wine_data)
wine_df.head()
```

```python
# get the types of the columns
wine_df.dtypes
```

```python
# get the column names
wine_df.columns
```

```python
# rename unamed to wine_id
wine_df = wine_df.rename(columns={"Unnamed: 0": "wine_id"})
wine_df.head()
```

```python
# set th eindex to wine_id
wine_df = wine_df.set_index("wine_id")
wine_df.head()
```

```python
# extract the year out of the title column
wine_year_df = wine_df.title.str.extract("(\d+)")
wine_year_df.head()
```

```python
# merge the new year df with the wine df
wine_df = pd.merge(wine_df, wine_year_df, how="inner", on="wine_id")
```

```python
# rename the column from 0 to year
wine_df = wine_df.rename(columns={0: "year"})
wine_df.head()
```

```python
# change the year column to a number
wine_df["year"] = pd.to_numeric(wine_df["year"])
```

```python
# verify the column types
wine_df.dtypes
```

```python
# using a dict and regex to rename some countries
dict1 = {
"US": "United States of America",
"England": "United Kingdom",
"Serbia": "Republic of Serbia"
}

# replacing the values
wine_df['country'] = wine_df['country'].replace(dict1)
wine_df.head()
```

```python
# get the original length of the database
len (wine_df)
```

```python
# drop all NANs or Nulls in the price column
wine_df = wine_df[pd.notnull(wine_df["price"])]
len(wine_df)
```

```python
# drop all the NANs or Nulls in the year column
wine_df = wine_df[pd.notnull(wine_df["year"])]
len(wine_df)
```

```python
# drop all the rows where the year is not over 1000
wine_df = wine_df[wine_df.year > 1000]
len(wine_df)
```

```python
# get the number of countries in df
wine_df["country"].nunique()
```

```python
# export it to a .csv file
wine_df.to_csv("../db/wine_data.csv", index = True)
```
